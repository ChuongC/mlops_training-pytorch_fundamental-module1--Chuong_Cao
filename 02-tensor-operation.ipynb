```python
import torch
torch.manual_seed(42)
```

```python
print("=" * 60)
print("Part 1: Basic Arithmetic Operations")
print("=" * 60)

# Create test tensors
x = torch.tensor([1, 2, 3, 4], dtype=torch.float32)
y = torch.tensor([5, 6, 7, 8], dtype=torch.float32)

print(f"x: {x}")
print(f"y: {y}")
```

```python
add_result = x + y
print(f"\nAddition (x + y): {add_result}")

#
sub_result = x - y
print(f"Subtraction (x - y): {sub_result}")

#
mul_result = x * y
print(f"Multiplication (x * y): {mul_result}")

#
div_result = x / y
print(f"Division (x / y): {div_result}")

#
pow_result = x ** 2
print(f"Power (x^2): {pow_result}")
```

```python
print("\n" + "=" * 60)
print("Part 2: Matrix Operations")
print("=" * 60)

# Create matrices
A = torch.randn(3, 4)
B = torch.randn(4, 5)

print(f"Matrix A shape: {A.shape}")
print(f"Matrix B shape: {B.shape}")

A, B, A.dtype, B.dtype, A.ndim, B.ndim
```

```python
#  Matrix multiplication
matmul_result = A @ B
print(f"\nMatrix multiplication A @ B shape: {matmul_result.shape}")

# Transpose
A_T = A.T
print(f"Transpose of A shape: {A_T.shape}")

#
min_val = torch.min(A)
max_val = torch.max(A)
print(f"\nMin value in A: {min_val:.4f}")
print(f"Max value in A: {max_val:.4f}")
```

```python
print("\n" + "=" * 60)
print("Part 3: Reduction Operations")
print("=" * 60)

# Create a 2D tensor
tensor = torch.randn(3, 4)
print(f"Tensor:\n{tensor}")
tensor.ndim
```

```python
sum_all = torch.sum(tensor)
print(f"\nSum of all elements: {sum_all:.4f}")

sum_rows = torch.sum(tensor, dim=0)
print(f"Sum along rows: {sum_rows}")

sum_cols = torch.sum(tensor, dim=1)
print(f"Sum along columns: {sum_cols}")

mean_all = torch.mean(tensor)
print(f"\nMean of all elements: {mean_all:.4f}")
#Standard deviation
std_all = torch.std(tensor)
print(f"Standard deviation: {std_all:.4f}")
```

```python
print("\n" + "=" * 60)
print("Part 4: Comparison Operations")
print("=" * 60)

x = torch.tensor([1,2,3,4,5])
y = torch.tensor([2, 2, 4, 4, 6])

print(f"x: {x}")
print(f"y: {y}")

# TODO: Element-wise comparison
eq_result = x == y
print(f"\nEqual (x == y): {eq_result}")

gt_result = x > y
print(f"Greater than (x > y): {gt_result}")

lt_result = x < y
print(f"Less than (x < y): {lt_result}")

clamped = torch.clamp(x, min=1, max=3)
print(f"\nClamp x to [1, 3]: {clamped}")
```

```python
print("\n" + "=" * 60)
print("Part 5: Broadcasting")
print("=" * 60)

x = torch.randn(4,1)
y = torch.randn(1,6)
print(f"x shape: {x.shape}")
print(f"y shape: {y.shape}")

result = x + y
print(f"\nBroadcast result shape: {result.shape}")
print(f"Result:\n{result}")
```

```python
# Exercise 2: Normalize a tensor to have zero mean and unit variance
print("\nExercise 2: Normalize tensor")
tensor = torch.randn(100, 10)
mean_tensor = torch.mean(tensor)
print(f"Mean of tensor: {mean_tensor:.4f}")
std_tensor = torch.std(tensor)
print(f"Standard deviation of tensor: {std_tensor:.4f}")
normalized_tensor = (tensor - mean_tensor) / std_tensor
print(f"\nNormalized tensor has {normalized_tensor.ndim} dimension and the shape: {normalized_tensor.shape}")
```

```python
# Exercise 3: Find the indices of the top 3 values in a tensor
print("\nExercise 3: Find top 3 indices")

tensor = torch.randn(10)

values, indices = torch.topk(tensor, k=3)
print(f"Original tensor: {tensor}")
print(f"Top 3 values: {values}")
print(f"Top 3 indices: {indices}")
```

```python
# Exercise 4: Implement softmax function
print("\nExercise 4: Implement softmax")
logits = torch.randn(2, 5)
#makes postive and amplifies diff
exp_logits = torch.exp(logits)
print(f"Logits:\n{logits}")
print(f"Exp Logits:\n{exp_logits}")
#Probabilities
softmax = exp_logits / torch.sum(exp_logits, dim=1, keepdim=True)
print(f"\nSoftmax:\n{softmax}")

print(f"\nSum per row: {softmax.sum(dim=1)}")
#Verify the ans
verify = torch.nn.functional.softmax(logits, dim=1)
print(f"\nVerify Softmax:\n{torch.allclose(softmax, verify)}")
```

```python
# Exercise 5: Compute cosine similarity between two vectors
print("\nExercise 5: Cosine similarity")
v1 = torch.randn(100)
v2 = torch.randn(100)
print(f"Compute cosine")
cosine_similarity = torch.dot(v1, v2) / (torch.linalg.norm(v1) * torch.linalg.norm(v2))
print(f"Cosine similarity: {cosine_similarity:.4f}")
```
